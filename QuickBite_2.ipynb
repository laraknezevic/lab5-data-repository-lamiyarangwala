The position of a damped pendulum bob is given by

$x(t) = A e^{-t/(2\tau)} \sin(\omega t)$ 

The potential energy of the bob is given by $U = m g \Delta z \approx  m g \frac{1}{2}\frac{x^2}{L}$ and the kinetic energy of the bob is given by $E = \frac{1}{2} m v^2$



1. With $m = 500g$, $A = 100$ (mm), $\tau = 5$ (s), and $\omega$ = $\pi$, please plot the potential energy, the kinetic energy, and the total energy (kinetic plus potential) of the bob from 0 to 15 seconds. All energies should have units of Joules (hint: this implies mass in kg and length in meters). Label the axes and add a legend.

1. On a second plot, make a semilog plot  (time axis linear, energy axis logarithmic) of the total energy, and the functions $U(0)*e^{-t/\tau}$ and $U(0)e^{-t/(2\tau)}$, where $U(0)$ is the initial total energy. Set the y-axis limits to the minimum and maximum of the total energy.

Hints: 
1. If you know that $\omega = \sqrt\frac{g}{L}$, you do not need to know $g$ or $L$ to do this problem. 
2. You can calculate the velocity analytically, or by using dx/dt = `np.gradient(x)/np.gradient(t)` with a fine time bin. 

```python
import numpy as np
import matplotlib.pyplot as plt

A = 100e-3
m = 500e-3
tau = 5
omega = np.pi
g = 9.81
L = g/omega**2
stepSize = 0.01


t = np.arange(0,15,stepSize)
x = A*np.exp(-t/2/tau)*np.sin(omega*t)
v = -A/2/tau*np.exp(-t/2/tau)*np.sin(omega*t) + omega*A*np.exp(-t/2/tau)*np.cos(omega*t)

U = m*g*0.5*x**2/L
KE = 0.5*m*v**2
E = KE + U
```

```python
plt.plot(t,KE, label = "KE")
plt.plot(t,U, label = "U")
plt.plot(t,E, label = "Total")
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
```

```python
U0 = E[0]
plt.semilogy(t, E, label = 'total')
plt.semilogy(t, U0*np.exp(-t/5), label = "U0*np.exp(-t/5)")
plt.semilogy(t, U0*np.exp(-t/10), label = "U0*np.exp(-t/10)")
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
```
